```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
sofia = {"lat": 42.6977, "lng": 23.3219}

bulgaria_map = Basemap(projection = "merc", llcrnrlat = 41, llcrnrlon = 22, urcrnrlat = 44.5, urcrnrlon = 29, resolution = "i")
bulgaria_map.drawcoastlines()
bulgaria_map.drawcountries()
x, y = bulgaria_map(sofia["lng"], sofia["lat"])
plt.plot(x, y, "p")
plt.show()
```
