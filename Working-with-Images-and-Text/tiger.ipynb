```python
from skimage.io import imread, imshow
import matplotlib.pyplot as plt
import numpy as np

from scipy.ndimage.filters import convolve
```

```python
tiger_img = imread("images/tiger.jpg")
```

```python
plt.imshow(tiger_img)
```

```python
red = tiger_img[:,:,0]
green = tiger_img[:,:,1]
blue = tiger_img[:,:,2]
red
```

```python
red.ravel() # make the matrix array 1 dimentional
```

```python
plt.hist(red.ravel(), bins = 256, color = "purple")
plt.show()
```

```python
plt.hist(red.ravel(), bins = 256, color = "red", alpha = 0.4)
plt.hist(green.ravel(), bins = 256, color = "yellow", alpha = 0.4)
plt.hist(blue.ravel(), bins = 256, color = "blue", alpha = 0.4)
plt.show()
```

```python
tiger_in_gray = np.mean(tiger_img, axis=2)
plt.imshow(tiger_in_gray)
```

```python
tiger_grayscale = 0.299 * red + 0.587 * green + 0.114 * blue 
plt.imshow(tiger_grayscale)
```

```python
box_blur_kernel = np.array([
[1, 1, 1],
[1, 1, 1],
[1, 1, 1]
]) / 9
```

```python
blurred = convolve(tiger_grayscale, box_blur_kernel)
```

```python
plt.imshow(tiger_grayscale[150:250, 300:400], cmap = "gray")
plt.show()
plt.imshow(blurred[150:250, 300:400], cmap = "gray")
plt.show()
```
