```python
from collections import Counter
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import *
```

```python
text = ""
with open("data/alice.txt", "r", encoding="UTF-8") as f:
    text = f.read()
print(len(text))
```

```python
chars_counter = Counter(text)
```

```python
chars_counter.most_common(5)
```

```python
words_countes = Counter(re.split("\\W+", text))
```

```python
words_countes.most_common(5)
```

```python
stops = set(stopwords.words("english"))
```

```python
text_words = re.split("\\W+", text.lower())
[w for w in text_words if w not in stops]
```

```python
Counter([w for w in text_words if w not in stops]).most_common(5)
```

```python
stemmer = PorterStemmer()
words = ["caresses", "flies", "dies", "seizing", "itemization", "sensational", "traditional", "reference", "plotted", "named"]
```

```python
print([stemmer.stem(w) for w in words])
```
