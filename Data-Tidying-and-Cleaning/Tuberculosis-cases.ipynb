```python
import pandas as pd
```

```python
tb = pd.read_csv("data/tb.csv")
tb.head(4)
```

```python
# Melt the values
tb_tidy = pd.melt(tb, id_vars = ["iso2","year"], 
             value_name = "cases", 
             var_name = "sex_and_age")
```

```python
# Separate the columns and merge back
partial_data = tb_tidy["sex_and_age"].str.extract("(\D)(\d+)(\d{2})", expand = True)
partial_data.columns = ["sex", "age_min", "age_max"]
partial_data["age"] = partial_data["age_min"] + "-" + partial_data["age_max"]
tb_tidy = pd.concat([tb_tidy, partial_data], axis = 1)
```

```python
# Remove unnessesary columns
tb_tidy = tb_tidy.drop(["sex_and_age", "age_min", "age_max"], axis = 1)
```

```python
# Remove missing values
tb_tidy = tb_tidy.dropna()
```

```python
# Sort them all
tb_tidy = tb_tidy.sort_values(by = ["iso2", "year", "sex", "age", "cases"])
```
