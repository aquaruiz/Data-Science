```python
junk_foods = pd.DataFrame([('ice', 51),
 ('pizza', 40),
 ('chocolate', 34),
 ('chips', 34),
 ('cheese', 23)], columns={"food", "qnty"})
```

```python
def get_deseased_count_by_food(food):
    return disease_data[disease_data.Food.str.lower().str.contains(food)== True].mean()[1]
```

```python
def get_fatalities_by_food(food):
    return disease_data[disease_data.Food.str.lower().str.contains(food)== True].mean().Fatalities
```

```python
current_list = []
for i in junk_foods.food:
    current_list.append([i, get_deseased_count_by_food(i), get_fatalities_by_food(i)])
```

```python
new_data = pd.DataFrame(current_list)
new_data.columns = ["food", "diseases", "fatalities"]
new_data
```

```python
new_table = pd.concat([junk_foods, new_data], axis = 1)
new_table.index = new_table.food
new_table.drop("food", axis = 1)
```

```python
new_table.corr()
```

```python
new_table["sugars"] = [12.008958,
2.756942,
31.747232,
8.148497,
4.846845]
```

```python
cat1 = new_table['sugars']
cat2 = new_table['diseases']

ttest_ind(cat1, cat2)
```

```python
stats.f_oneway(cat1, cat2)
```

```python
sugars_rating = pd.crosstab(new_table.diseases, new_table['sugars'])
sugars_rating
```

```python
chi2, pvalue, dof, expected = st.chi2_contingency(sugars_rating)
pvalue
```
